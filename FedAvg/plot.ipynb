```python
import os
import matplotlib.pyplot as plt
```

# 1. 联邦学习参与率$C$研究

### Non IID 

```python
log_dir = "./logs"
files = ["niid_accuracy_cf0.1.txt", "niid_accuracy_cf0.2.txt", "niid_accuracy_cf0.5.txt", "niid_accuracy_cf1.0.txt"]
```

```python
plt.figure(figsize=(10,5))
for filename in files:
    with open(os.path.join(log_dir, filename), "r") as f:
        lines = f.readlines()
    acc = []
    for line in lines[:150]:
        acc.append(float(line.split(":")[-1].strip()))
    plt.scatter(range(len(acc)), acc, marker='.')
plt.legend(["C=0.1", "C=0.2", "C=0.5", "C=1.0"])
plt.xlabel("Communication rounds")
plt.ylabel("Accuracy")
plt.title("Loop-Simulated Federated Learning on Non-IID Data (100 Clients)")
plt.savefig('figures/Non-IID-Loop-cf.png', dpi=300)
```
